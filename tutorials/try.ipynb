```python
import pandas as pd
from sas7bdat import SAS7BDAT
pd.set_option('display.max_columns', 100)
# 读取 .sas7bdat 文件
with SAS7BDAT('mon_predictors_final.sas7bdat') as file:
    df = file.to_data_frame()
numeric_cols = df.select_dtypes(include='number').columns.difference(['stkcd'])

# 用每列的中位数填充缺失值（不包括 'stkcd' 列）
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())

df
```

```python
with SAS7BDAT('ann_predictors.sas7bdat') as file:
    df1 = file.to_data_frame()
numeric_cols = df1.select_dtypes(include='number').columns.difference(['stkcd'])

# 用每列的中位数填充缺失值（不包括 'stkcd' 列）
df1[numeric_cols] = df1[numeric_cols].fillna(df1[numeric_cols].median())

df1
```

```python
df1['year']=df1['year']+1
df=df.merge(df1,on=['Stkcd','year'],how='left')
df
```

```python
numeric_cols = df.select_dtypes(include='number').columns.difference(['stkcd'])

# 用每列的中位数填充缺失值（不包括 'stkcd' 列）
df[numeric_cols] = df[numeric_cols].fillna(df[numeric_cols].median())
```

```python
import numpy as np
import torch
import torch.nn as nn

# Set and print device.
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
print(device)
import sys
sys.path.append('..')
from compfs.metrics import accuracy, gsim, tpr_fdr#评分指标
from compfs.models import CompFS, TorchModel#模型
from compfs.thresholding_functions import make_lambda_threshold
from compfs.datasets import NumpyDataset
is_classification = False
```

```python
df['sy']=(df['prc'].shift(-1) > df['prc']).astype(int)
df.dropna(inplace=True)
```

```python
y = df['prc']

# 选择所有不包含 'pre' 的列作为特征变量 X
X = df.drop(columns=['sy','prc','Stkcd','RD_ratio', 'RD_personr', 'RD_investr'])
train_mask = df['year'] <= 2018
X_train = X[train_mask]
y_train = y[train_mask]

X_val = X[~train_mask]
y_val = y[~train_mask]
X_train = X_train.to_numpy()
y_train = y_train.to_numpy()
X_val = X_val.to_numpy()
y_val = y_val.to_numpy()
print(X_train.shape,y_train.shape,X_val.shape,y_val.shape)
```

```python
compfs_config = {
    "model": CompFS,
    "model_config": {
        "lr": 0.003,
        "lr_decay": 0.99,
        "batchsize": 50,
        "num_epochs": 14,
        "loss_func": nn.SmoothL1Loss(),
        "val_metric": accuracy,
        "in_dim": 125,
        "h_dim": 20,
        "out_dim": 1,
        "nlearners": 10,
        "threshold_func": make_lambda_threshold(0.5),
        "temp": 0.5,
        "beta_s": 4.5,
        "beta_s_decay": 0.99,
        "beta_d": 1.2,
        "beta_d_decay": 0.99,
    },
}

compfs_config["device"] = device
```

```python
train_data = NumpyDataset(X_train, y_train, classification=is_classification)
val_data = NumpyDataset(X_val, y_val, classification=is_classification)
model = TorchModel(compfs_config)
model.train(train_data, val_data)
```

```python
learnt_groups = model.get_groups()
for i in range(len(learnt_groups)):
    print("Group: {}, Features: {}".format(i + 1, learnt_groups[i]))
```
